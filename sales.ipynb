```python
x = []
y = []
z = []

# Loop through each row in DataFrame 'c'
for i, j in c.iterrows():
    # Find unique 'pizza_name_id' without NaN
    z1 = sales.loc[(sales['pizza_name'] == j['pizza_name']) & (sales['pizza_size'] == j['pizza_size']), 'pizza_name_id'].dropna().unique()
    
    # Append 'pizza_name' and 'pizza_size'
    x.append(j['pizza_name'])
    y.append(j['pizza_size'])
    
    # If z1 has values, take the first one and remove brackets
    if z1.size > 0:
        z.append(z1[0])  # Add first element from z1 (already without brackets)
    else:
        z.append(None)  # If no match, append None

```

```python
def clean_date(a):
    return a.replace('-', '/')
```

```python
sales['order_date']=sales['order_date'].apply(clean_date)
```

```python
sales['order_date'] = pd.to_datetime(sales['order_date'], dayfirst=True, errors='coerce')
```

```python
sales['month'] = pd.to_datetime(sales['order_date'], dayfirst=True, errors='coerce').dt.month
sales['month'].unique()
```

```python
un_mon = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
sales['month'] = sales['month'].map(un_mon)
```

```python
sales['order_time'] = pd.to_datetime(sales['order_time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
# Convert 'order_date' to datetime format
sales['order_date'] = pd.to_datetime(sales['order_date'], format='%Y-%m-%d')

# Extract 'day_of_week' and 'year' as new features
sales['day_of_week'] = sales['order_date'].dt.day_name()
sales['year'] = sales['order_date'].dt.year

# Checking for missing values
missing_values = sales.isnull().sum()

# Display the updated sales and missing values summary
sales.head(200)

```
